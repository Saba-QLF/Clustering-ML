```python
# This library generate isotropic Gaussian blobs for clustering

from sklearn.datasets import make_blobs
X, y = make_blobs(n_samples=900,
                  n_features=2,
                  centers=3,
                  cluster_std=0.5,
                  shuffle=True,
                  random_state=0)
```

```python
import matplotlib.pyplot as plt
plt.scatter(X[:, 0], X[:, 1], c='white', marker='o',
            edgecolors='black', s=50)
```

# time for implementing kmeans in sklearn:

```python
from sklearn.cluster import KMeans
# n_cluster == K   init == how initial centroids
# Number of times the k-means algorithm is run with different centroid seeds.

km = KMeans(n_clusters=3,
            init="random",
            n_init=10,
            max_iter=300,
            tol=1e-4,
            random_state=0)

y_km = km.fit_predict(X)
```

```python
y_km
```

```python
plt.scatter(X[:, 0], X[:, 1], c=y_km, marker='o',
            edgecolors='black', s=50)
plt.scatter(km.cluster_centers_[:, 0],
            km.cluster_centers_[:, 1],
            marker="*", c='red', 
            edgecolors='black', s=250)
plt.show()
```

other names of soft kmeans:
1.fuzzy kmeans
2.fuzzy clustering
3.fuzzy C-means or FCM

```python
# inertia == cost
print('Distortion: %.2f' % km.inertia_)
```

# Elbow Method

```python
from sklearn.cluster import KMeans
K = 10
costs_values = []
for k in range(1, K+1):
    km = KMeans(n_clusters=k,
               init='k-means++',
               n_init=10,
               random_state=0)
    km.fit(X)
    costs_values.append(km.inertia_)

plt.plot(range(1, K+1), costs_values, marker='o')
plt.xlabel("number of clusters")
plt.ylabel("cost")
plt.show()
```
